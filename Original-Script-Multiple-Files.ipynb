## TRANSFERRING FROM HDF5 TO EXCEL

## Start here and press shift+enter to run each cell.

```python
import pandas as pd
import os
```

#### Below: Change parameters for what data you want.

```python
folder_location = r'G:\tone.light v2 stragglers\Results'
output_location = r'C:\Users\roselab\Desktop\test'
data = ['worm_index', 'n_frames', 'length', 'backward_frequency', 'backward_distance', 'backward_time', 'paused_frequency', 'paused_time', 'forward_distance', 'forward_time', 'forward_frequency' ]
```

### Below: The actual code. Don't change it.

```python
def save_excel_file(file_location, output_location, name, data):
    f = pd.HDFStore(file_location)
    means = f.get('features_means')
    writer = pd.ExcelWriter(r'{}\{}.xlsx'.format(output_location, name))
    means[data].to_excel(writer, 'Sheet1')
    writer.save()

for i in os.listdir(folder_location):
    if i.endswith('feat_manual.hdf5'):
        file_location = os.path.join(folder_location, i)
        save_excel_file(file_location, output_location, i[:-17], data)
```
